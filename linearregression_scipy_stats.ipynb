```python
import numpy as np
from pylab import *
%matplotlib inline
```

```python
pageSpeed= np.random.normal(3,1,1000)
purchaseamount= 100- (pageSpeed+np.random.normal(5,0.5,1000))*3
```

```python
scatter(pageSpeed,purchaseamount)
```

```python
import scipy.stats as stats
```

```python
slope, intercept,r_value,p_value,std_err =stats.linregress(pageSpeed,purchaseamount)
```

```python
slope
```

```python
intercept
```

```python
r_value**2
```

```python
import matplotlib.pyplot as plt
```

```python
def predict(x):
    return (slope*x)+ intercept
```

```python
fitline= predict(pageSpeed)
```

```python
plt.scatter(pageSpeed,purchaseamount)
plt.plot(pageSpeed,fitline,c='r')
```
