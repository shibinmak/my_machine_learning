```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
column_names= ['user_id','item_id','rating','timestamp']
```

```python
data= pd.read_csv('u.data' , sep = '\t' , names= column_names)
```

```python
data.head()
```

```python
movie_title= pd.read_csv('Movie_Id_Titles')
```

```python
movie_title.head()
```

```python
data = pd.merge(data,movie_title,on='item_id')
```

```python
data.head()
```

```python
data['rating'].hist(bins=30)
```

```python
 data.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
data.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating = pd.DataFrame(data.groupby('title')['rating'].mean(), columns=['rating'])
```

```python
rating['num of rating'] = pd.DataFrame(data.groupby('title')['rating'].count())
```

```python
rating.head()
```

```python
fig,ax1 = plt.subplots(figsize=(10,6))
ax1.hist(rating['num of rating'],bins=70)
plt.show()
```

```python
fig,ax1 = plt.subplots(figsize=(10,6))
ax1.hist(rating['rating'],bins=70)
plt.show()
```

```python
sns.jointplot('rating','num of rating',rating,size=10)
```

```python
moviemat = data.pivot_table(index='user_id',columns='title',values='rating' )
```

```python
moviemat.head()
```

```python
rating.sort_values('num of rating',ascending=False)
```

```python
starwars_userrating = moviemat['Star Wars (1977)']
liarliar_uesrrating=moviemat['Liar Liar (1997)']
```

```python
starwars_userrating.head()
```

```python
similar_to_starwars=moviemat.corrwith(starwars_userrating)
```

```python
similar_to_liarliar=moviemat.corrwith(liarliar_uesrrating)
```

```python
corr_starwars= pd.DataFrame(similar_to_starwars,columns=['corelaton'])
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars=corr_starwars.sort_values('corelaton',ascending=False)
```

```python
corr_starwars = corr_starwars.join(rating['num of rating'])
```

```python
corr_starwars=corr_starwars[corr_starwars['num of rating']>100]
```

```python
corr_starwars.sort_values('corelaton',ascending=False)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['corealtion'])
```

```python
corr_liarliar.dropna(inplace=True)
```

```python
corr_liarliar=corr_liarliar.join(rating['num of rating'])
```

```python
corr_liarliar=corr_liarliar.sort_values('corealtion',ascending=False)
```

```python
corr_liarliar=corr_liarliar[corr_liarliar['num of rating']>100]
```

```python
corr_liarliar.sort_values('corealtion',ascending=False)
```
