```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(2)
pagespeed= np.random.normal(3.0,1.0,1000)
purchasedamount= np.random.normal(50.0,10.0,1000)/pagespeed
```

```python
plt.scatter(pagespeed,purchasedamount,marker='*')
```

```python
x=np.array(pagespeed)
y=np.array(purchasedamount)
p4= np.poly1d(np.polyfit(x,y,4))
```

```python
xp= np.linspace(0,7,1000)
```

```python
plt.scatter(pagespeed,purchasedamount)
plt.plot(xp,p4(xp),c='r')
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, p4(x))

print(r2)
```
