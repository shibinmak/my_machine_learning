```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data= make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
plt.scatter(data[0][:,0],data[0][:,1])
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    knni= KMeans(n_clusters=i)
    knni.fit(data[0])
    wcss.append(knni.inertia_)
    
```

```python
plt.title('K value estimation')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.plot(range(1,11),wcss,linestyle='dashed',color='red',marker='*',markersize='15',markerfacecolor='black')
```

```python
classifier = KMeans(n_clusters=3)
```

```python
classifier.fit(data[0])
```

```python
classifier.cluster_centers_
```

```python
classifier.labels_
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.scatter(data[0][:,0],data[0][:,1],c=classifier.labels_ , cmap='rainbow')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1] , cmap='rainbow')
ax1.set_title('K-MEANS')
ax2.set_title('original')
```

```python
fig,(ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.scatter(data[0][:,0],data[0][:,1],c=classifier.labels_ , cmap='rainbow')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1] , cmap='rainbow')
```
